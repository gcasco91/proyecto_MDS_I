## Exploracion y transformacion

```python
# Normalizar nombres de columnas
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace(r"[치칠칤칩칰침]", lambda m: {
        "치": "a", "칠": "e", "칤": "i", "칩": "o", "칰": "u", "침": "n"
    }[m.group()], regex=True)
)
```

```python
df.columns.tolist()  # Ver para confirmar que quedaron limpias
```

```python
# Detectar columnas que contienen 'fecha' o 'dia' en el nombre
columnas_fecha = [col for col in df.columns if 'fecha' in col or 'dia' in col]
print("Columnas con fechas detectadas:", columnas_fecha)
```

```python
# Convertir a datetime en formato d칤a/mes/a침o (formato habitual en Excel en espa침ol)
for col in columnas_fecha:
    df[col] = pd.to_datetime(df[col], format="%d/%m/%Y", errors="coerce")

# Verificamos los tipos de datos
df[columnas_fecha].dtypes
```

```python
# Filtrar registros con inconsistencias
incoherentes_cierre_vs_apertura = df[
    (df['ace_fecha_cierre'].notnull()) &
    (df['dia_apertura'].notnull()) &
    (df['ace_fecha_cierre'] < df['dia_apertura'])
]

print(f"Registros con fecha de cierre anterior a apertura: {incoherentes_cierre_vs_apertura.shape[0]}")
```

```python
incoherentes_primer_cierre = df[
    (df['fecha_pimer_cierre'].notnull()) &
    (df['dia_apertura'].notnull()) &
    (df['fecha_pimer_cierre'] < df['dia_apertura'])
]

print(f"Registros con primer cierre antes de la apertura: {incoherentes_primer_cierre.shape[0]}")
```

```python
# Conteo de nulos y porcentaje
nulos = df.isnull().sum().to_frame(name='nulos')
nulos['%_nulos'] = (nulos['nulos'] / len(df)) * 100
nulos = nulos[nulos['nulos'] > 0].sort_values(by='%_nulos', ascending=False)

nulos.head(20)
```

```python
import matplotlib.pyplot as plt

nulos.head(20).sort_values('%_nulos').plot(
    kind='barh',
    figsize=(10, 8),
    title='Top 20 columnas con m치s valores nulos',
    legend=False
)
plt.xlabel('valores nulos')
plt.tight_layout()
plt.show()
```

```python
# Selecci칩n de columnas categ칩ricas
cat_cols = df.select_dtypes(include='object').columns

# Tomamos las primeras columnas con baja cardinalidad (menos de 20 categor칤as)
for col in cat_cols:
    if df[col].nunique() < 20:
        plt.figure(figsize=(10, 4))
        sns.countplot(data=df, x=col, order=df[col].value_counts().index, palette='viridis')
        plt.title(f'Distribuci칩n de la variable: {col}')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
```

```python
# Filtrar registros donde dm  "CUSTOMER_SERVICE"
if 'dm' in df.columns:
    df = df[df['dm'] != "CUSTOMER_SERVICE"]
```

```python
# Filtrar registros donde Op_estado es "Cerrado" (estado para operaciones) y Ace_estado es "Abierto" (estado para sistema de gestion)
condicion = (df['op_estado'] == 'Cerrado') & (df['ace_estado'] == 'Abierto')
casos_inconsistentes = df[condicion]

print(f"Total de registros con Op_estado='Abierto' y Ace_estado='Cerrado': {len(casos_inconsistentes)}")
casos_inconsistentes.head()
```

```python
# Eliminar registros inconsistentes: Op_estado = "Cerrado" y Ace_estado = "Abierto"
df = df[~((df['op_estado'] == 'Cerrado') & (df['ace_estado'] == 'Abierto'))]
```

```python
# Eliminar registros donde dm = "CUSTOMER_SERVICE" -> para analisis de pagos/rechazos no es relevante
df = df[df['dm'] != 'CUSTOMER_SERVICE']
```

```python
#estado actual de los registros
print(f"Total de registros despu칠s de eliminar inconsistencias: {len(df)}")
```

```python
print("Registros antes del filtro de fecha:", len(df))
print("Registros con fecha_apertura < 2024-01-01:", (df['dia_apertura'] < '2024-01-01').sum())
```

```python
#An치lisis de importes reclamados vs. pagados
print("Importe reclamado:")
print(df['importe_reclamado_hojagastos'].describe())

print("\nImporte pagado:")
print(df['importe_indemnizado_hojagastos'].describe())
```

```python
# revisar si hay valores negativos en los importes
print("Valores negativos en importe reclamado:")
print((df['importe_reclamado_hojagastos'] < 0).sum())
print("Valores negativos en importe pagado:")
print((df['importe_indemnizado_hojagastos'] < 0).sum())
```

```python
# Eliminar registros con importes negativos
df = df[(df['importe_reclamado_hojagastos'] >= 0) & 
        (df['importe_indemnizado_hojagastos'] >= 0)]

```

```python
# detectar y eliminar outliers usando el m칠todo IQR (importe reclamado)
Q1 = df['importe_reclamado_hojagastos'].quantile(0.25)
Q3 = df['importe_reclamado_hojagastos'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df['importe_reclamado_hojagastos'] >= lower_bound) & (df['importe_reclamado_hojagastos'] <= upper_bound)]

```

```python

# Verificar si hay outliers despu칠s de la eliminaci칩n
print("Registros despu칠s de eliminar outliers:", len(df))
```

```python
# detectar y eliminar outliers usando el m칠todo IQR (importe pagado)
Q1 = df['importe_indemnizado_hojagastos'].quantile(0.25)
Q3 = df['importe_indemnizado_hojagastos'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df['importe_indemnizado_hojagastos'] >= lower_bound) & (df['importe_indemnizado_hojagastos'] <= upper_bound)]

```

```python
#tama침o del dataframe limpio
print(f"Total de registros despues de limpieza: {len(df)}")

```

```python
# Tasa por pa칤s (si existe la columna)
if 'pais_origen_pl_prod' in df.columns:
    df.groupby('pais_origen_pl_prod')['tasa_pago'].mean().sort_values(ascending=False).plot(kind='bar', figsize=(12,5))
    plt.title('Tasa media de pago por pa칤s')
    plt.ylabel('Tasa de pago')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()


```
